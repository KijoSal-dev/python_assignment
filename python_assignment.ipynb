```python
#6)Create a tuple and print its second and last elements.

my_list = [100, 85, 70, 51, 40, 30, 15]
my_tuple = tuple(my_list)
print(my_tuple[1])
print(my_tuple[-1])
```

```python
#7)Convert a tuple into a list.

my_tuple = (10,40, 60, 80, 100)
my_list = list(my_tuple)
print(my_list)
```

```python
#8)Check whether a number is positive, negative, or zero using if-elif-else.

number = -50

if number > 0:
    print("Positive")
if number < 0:
  print("Negative")
else:
  print("The number is zero")
```

```python
#9)Check whether a number is even or odd.

number = 75

if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#10)Use a for loop to print numbers from 1 to 20.

for i in range(1,21):
     print(i)
```

```python
#11)Use a while loop to print numbers from 10 to 1.

i = 10
while i >= 1:
    print(i)
    i -= 1
```

```python
#12)Find the sum of all elements in a list using a loop.

my_list = [10, 20, 30, 40, 80]
sum = 0
for i in my_list:
    sum += i
print(sum)
```

```python
#13)Print all numbers divisible by 3 from a list.

my_list = [30, 45, 3, 9, 20, 50, 60, 90, 99, 5, 55, 34]
div = 3
for i in my_list:
  if i % div == 0:
    print(i)
```
